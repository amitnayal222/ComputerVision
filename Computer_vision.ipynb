```python
import cv2
import imutils
import sys
```

```python


cascPath = sys.argv[1]
faceCascade = cv2.CascadeClassifier("C:/Users/600034724/Desktop/Learning/FaceDetect-master/FaceDetect-master/haarcascade_frontalface_default.xml")

video_capture = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = faceCascade.detectMultiScale(
        image = gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30),
        flags=cv2.CASCADE_SCALE_IMAGE
    
    )

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 2)

    # Display the resulting frame
    cv2.imshow('Video', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
# When everything is done, release the capture
video_capture.release()
cv2.destroyAllWindows()
```

```python

cascPath = sys.argv[1]
faceCascade = cv2.CascadeClassifier("C:/Users/600034724/Desktop/Learning/FaceDetect-master/FaceDetect-master/haarcascade_frontalface_default.xml")
eyeCascade = cv2.CascadeClassifier("C:/Users/600034724/Desktop/Learning/opencv-master/opencv-master/data/haarcascades/haarcascade_eye.xml")
smileCascade = cv2.CascadeClassifier("C:/Users/600034724/Desktop/Learning/opencv-master/opencv-master/data/haarcascades/haarcascade_smile.xml")
bodyCascade = cv2.CascadeClassifier("C:/Users/600034724/Desktop/Learning/opencv-master/opencv-master/data/haarcascades/haarcascade_fullbody.xml")
halfCascade = cv2.CascadeClassifier("C:/Users/600034724/Desktop/Learning/opencv-master/opencv-master/data/haarcascades/haarcascade_upperbody.xml")
# faceCascade = cv2.CascadeClassifier(cascPath)
# eyeCascade = cv2.CascadeClassifier(eyePath)
# smileCascade = cv2.CascadeClassifier(smilePath)
video_capture = cv2.VideoCapture(0)
font = cv2.FONT_HERSHEY_SIMPLEX
while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()
    frame = imutils.resize(frame, width=900) # resize original video for better viewing performance
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)    
    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=25,
        minSize=(30, 30),
        flags=cv2.CASCADE_SCALE_IMAGE
        )
    for (x, y, w, h) in faces:
        if w < 250 :
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = frame[y:y+h, x:x+w]
    smile = smileCascade.detectMultiScale(
        roi_gray,
        scaleFactor= 1.16,
        minNeighbors=35,
        minSize=(25, 25),
        flags=cv2.CASCADE_SCALE_IMAGE
        )
    for (sx, sy, sw, sh) in smile:
        cv2.rectangle(roi_color, (sh, sy), (sx+sw, sy+sh), (0, 0, 255), 2)
        cv2.putText(frame,'Smile',(x + sx,y + sy), 1, 1, (0, 255, 0), 1)

    eyes = eyeCascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        cv2.putText(frame,'Eye',(x + ex,y + ey), 1, 1, (0, 255, 0), 1)
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # convert video to grayscale
    half_body = halfCascade.detectMultiScale(
        gray,
        scaleFactor=1.07,
        minNeighbors=5,
        minSize=(30,90),
        flags=cv2.CASCADE_SCALE_IMAGE
        )
    for (bx,by,bw,bh) in half_body:
        cv2.rectangle(frame, (bx,by), (bx+bw,by+bh), (0, 0, 255), 3)
        cv2.putText(frame,'Half_Body',(x+bx,y+by), 1, 1, (0, 0, 255), 2)
    body = bodyCascade.detectMultiScale(
        gray,
        scaleFactor=1.03,
        minNeighbors=4,
        minSize=(60,130),
        flags=cv2.CASCADE_SCALE_IMAGE
        )
    for (bx,by,bw,bh) in body:
        cv2.rectangle(frame, (bx,by), (bx+bw,by+bh), (0, 0, 255), 3)
        cv2.putText(frame,'Body',(x+bx,y+by), 1, 1, (0, 0, 255), 2)

    cv2.putText(frame,'Number of Faces : ' + str(len(faces)),(30, 30), font, 1,(255,0,0),2)      
        # Display the resulting frame
    cv2.imshow('Video', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video_capture.release()
cv2.destroyAllWindows()
```
